```python
"""
Ищу выбросы на основе документации и dataframe.describe().
Согласно документации:  
 - surgery 1 или 2. Вывод - выбросов нет
 - age 1 или 2. Показывает, что max 9. Посмотрю ниже сколько таких показателей
 - rectal_temp. Показатели в пределах нормы
 - pulse. Есть max показатель 184, что больше нормы по документации, надо проверить сколько таких высоких показателей
 - resp_rate. Есть max 96, по документации норма от 8 до 10. Проверю ниже
 - temp_extr от 1 до 4. Выбросов нет
 - pain от 1 до 5. Выбросов нет
 - outcome 1 / 2 / 3. Выбросов нет
 
"""
```

```python
# AGE 1 / 2. Все выбросы. Вместо 1 или 2, появилась 9. Пока ничего с ними не делаю

print(df_columns[(df_columns['age'] < 1) | (df_columns['age'] > 4)])
```

```python
# Ищу максимальные выбросы через межквартильный размах. 3 квартиль = 148. Вывожу, все что больше. Но оставлю,
# такие показатели могут быть

q1 = df_columns['pulse'].quantile(0.25)
q3 = df_columns['pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)

print (upper_bound)
print(df_columns[(df_columns['pulse'] > 148.0)])
```

```python
# Ищу максимальные выбросы через межквартильный размах. 3 квартиль = 62.25. Вывожу, все что больше. 
# Пока оставлю

q1 = df_columns['resp_rate'].quantile(0.25)
q3 = df_columns['resp_rate'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)

print (upper_bound)
print(df_columns[(df_columns['resp_rate'] > 63.0)])
```

### Задание 3. Работа с пропусками

```python
df_columns.isna().sum()
```

```python
"""
Проверяю 1-ый пропуск он есть 'surgery' и в 'outcome' .
То есть нет инф-ции по операции и по результату. 
Так как пропуск всего 1, я считаю, что можно удалить эту строку,
все остальные показатели среднии и не приведут к изменениям глобальным
Запишу изменения в новый dataframe (df_horse)

"""
df_columns[df_columns['surgery'].isnull()]
```

```python
df_horse = df_columns[df_columns.surgery.notnull()]
```

```python
"""
Все остальные пропуски я заменила средним, так как не вижу связи между показателями
"""
```

```python
df_horse['rectal_temp'].fillna(df_columns['rectal_temp'].mean(), inplace=True)
```

```python
df_horse['pulse'].fillna(df_columns['pulse'].mean(), inplace=True)
```

```python
df_horse['resp_rate'].fillna(df_columns['resp_rate'].mean(), inplace=True)
```

```python
df_horse['temp_extr'].fillna(df_columns['temp_extr'].mean(), inplace=True)
```

```python
df_horse['pain'].fillna(df_columns['pain'].mean(), inplace=True)
```
